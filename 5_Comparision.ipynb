```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
x = ['Decisionn Tree', 'Random Forest', 'Bagging', 'Adaptive Boosting', 'Gradient Boosting']
r_squares = [50, 49, 91, 31, 37]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, r_squares, color='green')
plt.xticks(rotation=90)
plt.ylabel("R-Square")
plt.title("Performance")
plt.xticks(x_pos, x)
plt.show()
```

```python
import numpy as np
N = 5
train_rmse = (42.0, 39.76, 15.98, 44.74, 42.57)
test_rmse = (37.84, 38.29, 41.60, 44.50, 42.56)
ind = np.arange(N) 
width = 0.35       
plt.bar(ind, train_rmse, width, label='Train RMSE')
plt.bar(ind + width, test_rmse, width,
    label='Test RMSE')
plt.ylabel('RMSE')
plt.title('Fitting')
plt.xticks(ind + width / 2, ('Decisionn Tree', 'Random Forest', 'Bagging', 'Adaptive Boosting', 'Gradient Boosting'))
plt.xticks(rotation=90)
plt.legend(loc='best')
plt.show()
```

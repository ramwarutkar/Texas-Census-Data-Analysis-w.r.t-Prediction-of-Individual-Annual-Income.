```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Proccessed_Texas_Employment_ACS_2015_Texas.csv',header=0)
df.head()
```

```python
df.dropna()
df.info()
```

```python
# To convert categorial variable to numeric
# Using dummy variable
# Number of dummy variables = Number of category - 1
# Dummies a pandas method
# drop_first to have number of categorical variable - 1
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
# for splitting data 
X=df.loc[:,df.columns!='WAGP']
y=df['WAGP']
from sklearn.model_selection import train_test_split 
# If randam_state is constant we get the same split every time
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Training Decision Tree
from sklearn import tree
regtree=tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(X_train,y_train)
y_train_pred=regtree.predict(X_train)
y_test_pred=regtree.predict(X_test)
```

```python
# Model Performance
from sklearn.metrics import mean_squared_error,r2_score
print('RMSE - ',(mean_squared_error(y_test,y_test_pred))**0.5) 
print('R2 Train - ',r2_score(y_train,y_train_pred)) # Goodness of fit 0 to 1
print('R2 Test - ',r2_score(y_test,y_test_pred))
```

```python
# Plotting decision tree
# First make a dot file then convert the dot file into image, then use this 
# image to display the plot 
dot_data=tree.export_graphviz(regtree,feature_names=X_train.columns,out_file=None)
from IPython.display import Image
import pydotplus
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
rf_reg=RandomForestRegressor(n_estimators=1000,n_jobs=-1,random_state=42)
rf_reg.fit(X_train,y_train)
y_train_rf_pred=rf_reg.predict(X_train)
y_test_rf_pred=rf_reg.predict(X_test)
print('RMSE - ',(mean_squared_error(y_test,y_test_rf_pred))**0.5) 
print('R2 Train - ',r2_score(y_train,y_train_rf_pred)) # Goodness of fit 0 to 1
print('R2 Test - ',r2_score(y_test,y_test_rf_pred))
```

```python
# Optimizing Hyperparameter values
from sklearn.model_selection import GridSearchCV
rf_reg=RandomForestRegressor(n_estimators=250,random_state=42)
param_grid={'max_features':[4,5,6,7,8,9,10],
            'min_samples_split':[2,3,10]}
grid_search=GridSearchCV(rf_reg,param_grid,n_jobs=-1,cv=5)
grid_search.fit(X_train,y_train)
print(grid_search.best_params_)
cv_rf_reg=grid_search.best_estimator_
y_train_rf_cv_pred=cv_rf_reg.predict(X_train)
y_test_rf_cv_pred=cv_rf_reg.predict(X_test)
print('RMSE - ',(mean_squared_error(y_test,y_test_rf_cv_pred))**0.5) 
print('R2 Train - ',r2_score(y_train,y_train_rf_cv_pred)) # Goodness of fit 0 to 1
print('R2 Test - ',r2_score(y_test,y_test_rf_cv_pred))
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
gbc_reg=GradientBoostingRegressor(learning_rate=0.02,n_estimators=1000,max_depth=1)
gbc_reg.fit(X_train,y_train)
y_train_gbc_pred=gbc_reg.predict(X_train)
y_test_gbc_pred=gbc_reg.predict(X_test)
print('RMSE - ',(mean_squared_error(y_test,y_test_gbc_pred))**0.5) 
print('R2 Train - ',r2_score(y_train,y_train_gbc_pred)) # Goodness of fit 0 to 1
print('R2 Test - ',r2_score(y_test,y_test_gbc_pred))
```

```python
# Adaptive Boosting
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error,r2_score
ada_reg=AdaBoostRegressor(learning_rate=0.02,n_estimators=5000)
ada_reg.fit(X_train,y_train)
y_train_ada_pred=ada_reg.predict(X_train)
y_test_ada_pred=ada_reg.predict(X_test)
print('RMSE - ',(mean_squared_error(y_test,y_test_ada_pred))**0.5) 
print('R2 Train - ',r2_score(y_train,y_train_ada_pred)) # Goodness of fit 0 to 1
print('R2 Test - ',r2_score(y_test,y_test_ada_pred))
```

```python
# XG Boost
import xgboost as xgb
from sklearn.metrics import mean_squared_error,r2_score
xgb_reg=xgb.XGBRegressor(max_depth=5,n_estimators=10000,learning_rate=0.3,n_jobs=-1)
xgb_reg.fit(X_train,y_train)
y_train_xgb_pred=xgb_reg.predict(X_train)
y_test_xgb_pred=xgb_reg.predict(X_test)
print('RMSE - ',(mean_squared_error(y_test,y_test_xgb_pred))**0.5) 
print('R2 Train - ',r2_score(y_train,y_train_xgb_pred)) # Goodness of fit 0 to 1
print('R2 Test - ',r2_score(y_test,y_test_xgb_pred))
xgb.plot_importance(xgb_reg) # Graph of relative importance of variable in the model
```

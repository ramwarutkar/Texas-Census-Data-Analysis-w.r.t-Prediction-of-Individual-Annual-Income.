```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Proccessed_Texas_Employment_ACS_2015_Texas.csv',header=0)
df.head()
```

```python
df.dropna()
df.info()
```

```python
# To convert categorial variable to numeric
# Using dummy variable
# Number of dummy variables = Number of category - 1
# Dummies a pandas method
# drop_first to have number of categorical variable - 1
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
# for splitting data 
X=df.loc[:,df.columns!='WAGP']
y=df['WAGP']
from sklearn.model_selection import train_test_split 
# If randam_state is constant we get the same split every time
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVR
```

```python
# rbf SVM
from sklearn.svm import SVR
rbf_svm=SVR(kernel='rbf',C=1000)
rbf_svm.fit(X_train,y_train)
y_train_pred=rbf_svm.predict(X_train,y_train)
y_test_pred=rbf_svm.predict(X_test,y_test)
# Linear Model Performance
from sklearn.metrics import mean_squared_error,r_score
print('Train MSE -',mean_squared_error(y_train,y_train_pred))
print('Test MSE - ',mean_squared_error(y_test,y_test_pred))
print('Train R2 - ',r2_score(y_train,y_train_pred))
print('Test R2 - ',r2_score(y_test,y_test_pred))
```

```python
# Hyper parameter tunning - Kernel - Linear 
from sklearn.model_selection import GridSearchCV
linear_svm=SVR(kernel='linear')
param_grid={'C':[0.1,0.5,1,2,3,4,5]}
linear_svm_grid_cv=GridSearchCV(linear_svm,param_grid,n_jobs=-1,cv=3,verbos=-1)
linear_svm_grid_cv.fit(X_train,y_train)
linear_svm=linear_svm_grid_cv.best_params_
y_train_pred=linear_svm.predict(X_train,y_train)
y_test_pred=linear_svm.predict(X_test,y_test)
# Linear Model Performance
from sklearn.metrics import mean_squared_error,r_score
print('Train MSE -',mean_squared_error(y_train,y_train_pred))
print('Test MSE - ',mean_squared_error(y_test,y_test_pred))
print('Train R2 - ',r2_score(y_train,y_train_pred))
print('Test R2 - ',r2_score(y_test,y_test_pred))
```
